```python
#Importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Load the image using OpenCV
img = cv2.imread("download.jpeg")
# Convert BGR (OpenCV's default color format) to RGB (Matplotlib's default color format)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Display the original image
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
# Convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Display the grayscale image
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

Inverting an image helps to enhance details or preparing images for further analysis.It converts  white pixels to black and vice versa in grayscale images

```python
# Invert the grayscale image
img_invert = cv2.bitwise_not(img_gray)
# Display the inverted image
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

Smoothing helps to remove noise and small fluctuations in intensity, resulting in a more uniform appearance and making subsequent processing steps more robust.
Gaussian blur is a versatile tool in image processing, commonly used for tasks like noise reduction, smoothing, edge detection, and feature extraction.

```python
# Apply Gaussian blur to smoothen the image
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
# Display the smoothened image
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

Dividing the inverted grayscale image by the smoothened image effectively amplifies the intensity differences along the edges, resulting in a more pronounced sketch-like appearance.

```python
# Create the final sketch image by dividing the inverted image by the smoothened image
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
# Display all images in a single row with titles
plt.figure(figsize=(20,20))
# Original Image
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
# Grayscale Image
plt.subplot(1,5,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
#Inverted Image
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
#Smoothened Image
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
#Final Sketch
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
